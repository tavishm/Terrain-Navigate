```python
import os
import xml.etree.ElementTree as ET

# Define the directory path
metadata_dir = "/home/tavish/Documents/URC_Path_Finding/junk/data/Projects/USGS_LPC_UT_Southern_QL1_2018_LAS_2019/metadata/"

# Initialize the list to store parsed XML data
xml_data = {} # id: xml

# Get all XML files in the directory
xml_files = [f for f in os.listdir(metadata_dir) if f.endswith('.xml')]

print(f"Found {len(xml_files)} XML files")

# Load each XML file
for xml_file in xml_files:
    file_path = os.path.join(metadata_dir, xml_file)
    try:
        tree = ET.parse(file_path)
        root = tree.getroot()
        xml_data[file_path] = root
    except Exception as e:
        print(f"Error loading {xml_file}: {e}")

print(f"Successfully loaded {len(xml_data)} XML files")
```

```python
import numpy as np

MDRS_coords = [-110.7919,38.4065]

ex_coords = {
    "NW": (38.433521, -110.826259),
    "NE": (38.433521, -110.757541),
    "SE": (38.379469, -110.757566),
    "SW": (38.379469, -110.826234)

}

candidates = []

for fpath, xml_o in xml_data.items():
    west = np.float128(xml_o[0][4][0][0].text)
    east = np.float128(xml_o[0][4][0][1].text)
    north = np.float128(xml_o[0][4][0][2].text)
    south = np.float128(xml_o[0][4][0][3].text)

    if (west < ex_coords["NE"][1] and east > ex_coords["NW"][1]) and (south < ex_coords["NW"][0] and north > ex_coords["SW"][0]):
        #print("Found", east, west, north, south)
        candidates.append([fpath.replace("/home/tavish/Documents/URC_Path_Finding/junk/data/Projects/USGS_LPC_UT_Southern_QL1_2018_LAS_2019/metadata/USGS_LPC_UT_Southern_QL1_2018_", "").replace("_LAS_2019_meta.xml", ""), west, east, north, south])

print("Found candidates:", len(candidates))
```

```python
# Download candidate files

import urllib.request
parent_link = "https://rockyweb.usgs.gov/vdelivery/Datasets/Staged/Elevation/LPC/Projects/USGS_LPC_UT_Southern_QL1_2018_LAS_2019/laz/USGS_LPC_UT_Southern_QL1_2018_{}_LAS_2019.laz"

for candidate in candidates:
    file_id = candidate[0]
    file_url = parent_link.format(file_id)
    
    # Create candidates directory if it doesn't exist
    os.makedirs("candidates", exist_ok=True)
    
    # Check if file already exists
    filename = f"candidates/{file_id}.laz"
    if os.path.exists(filename):
        print(f"Skipping {file_id} - already downloaded")
        continue
    
    print("Downloading:", file_url)
    
    # Download the file
    urllib.request.urlretrieve(file_url, filename)
    print(f"  Saved to: {filename}")
```
